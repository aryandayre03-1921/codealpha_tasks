Exploratory Data Analysis (EDA)

1. How many movies are in the dataset?

```python
total_movies = df.shape[0]
print(f"Total number of movies: {total_movies}")
```

2. Which movies are the most popular?

```python
top_popular = df.sort_values(by="popularity", ascending=False).head(10)


print("Top 10 Most Popular Movies:")
display(top_popular[["title", "popularity"]])  


colors = list(plt.cm.viridis(np.linspace(0.2, 0.8, len(top_popular))))


plt.figure(figsize=(10,6))
sns.barplot(
    x="popularity",
    y="title",
    data=top_popular,
    hue="title",       
    palette=colors,
    dodge=False,       
    legend=False       
)
plt.title("Top 10 Most Popular Movies", fontsize=16, weight='bold')
plt.xlabel("Popularity Score")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```

3. Which movies have the highest ratings?

```python
top_rated = df.sort_values(by="vote_average", ascending=False).head(10)

print("Top 10 Highest Rated Movies:")

display(top_rated[["title", "vote_average"]])


colors = list(plt.cm.plasma(np.linspace(0.2, 0.8, len(top_rated))))


plt.figure(figsize=(10,6))
sns.barplot(
    x="vote_average",
    y="title",
    data=top_rated,
    hue="title",     
    palette=colors,
    dodge=False,
    legend=False
)
plt.title("Top 10 Highest Rated Movies", fontsize=16, weight='bold')
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```

4. What is the distribution of movie ratings?

```python
print("Distribution Of Movie Ratings:")

display(df[["title", "vote_average"]])


plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="vote_average",
    bins=20,
    kde=True,
    color=sns.color_palette("plasma", as_cmap=True)(0.6)  # Best gradient color
)
plt.title("Distribution of Movie Ratings", fontsize=16, weight='bold')
plt.xlabel("Average Rating")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```

5. Which movies generated the highest revenue?

```python
top_revenue = df.sort_values(by="revenue", ascending=False).head(10)

print("Top 10 highest revenue movies:")

display(top_revenue[["title", "revenue"]])


colors = list(plt.cm.inferno(np.linspace(0.2, 0.8, len(top_revenue))))


plt.figure(figsize=(10,6))
sns.barplot(
    x="revenue",
    y="title",
    data=top_revenue,
    hue="title",    
    palette=colors,
    dodge=False,
    legend=False
)
plt.title("Top 10 Highest Revenue Movies", fontsize=16, weight='bold')
plt.xlabel("Revenue")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```

6. What are the most common languages of movies?

```python
top_languages = df["original_language"].value_counts().head(10)

print("Top 10 Movie Languages:")
display(pd.DataFrame({
    "Language": top_languages.index,
    "Number of Movies": top_languages.values
}))

# Gradient colors
colors = list(plt.cm.magma(np.linspace(0.2, 0.8, len(top_languages))))

# Bar chart (Seaborn, warning-free)
plt.figure(figsize=(10,6))
sns.barplot(
    x=top_languages.values,
    y=top_languages.index,
    hue=top_languages.index,  # Assign hue to remove FutureWarning
    palette=colors,
    dodge=False,
    legend=False
)
plt.title("Top 10 Movie Languages", fontsize=16, weight='bold')
plt.xlabel("Number of Movies")
plt.ylabel("Language")
plt.tight_layout()
plt.show()
```

7. Is there a relationship between budget and revenue?

```python
budget_revenue_table = df[["title", "budget", "revenue", "popularity"]].sort_values(by="revenue", ascending=False).head(10)
display(budget_revenue_table)

# Bubble plot for Budget vs Revenue
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    x=df["budget"],
    y=df["revenue"],
    s=df["popularity"]*5,            # Bubble size proportional to popularity
    c=df["revenue"],                 # Color gradient based on revenue
    cmap="plasma",                    # Attractive professional gradient
    alpha=0.7,
    edgecolor="k"
)

plt.xscale("log")  # Use log scale for better visualization
plt.yscale("log")
plt.colorbar(scatter, label="Revenue")
plt.title("Budget vs Revenue of Movies", fontsize=16, weight='bold')
plt.xlabel("Budget (log scale)")
plt.ylabel("Revenue (log scale)")
plt.tight_layout()
plt.show()
```

8. What is the distribution of top 10 movie runtimes?

```python
top_10_runtime = df[['title', 'runtime']].sort_values(by='runtime', ascending=False).head(10)
display(top_10_runtime)

# Plot histogram with hue for better clarity
plt.figure(figsize=(10,6))
sns.histplot(
    df, 
    x="runtime", 
    bins=30, 
    hue=pd.cut(df['runtime'], bins=5),  # Categorize for hue
    palette="Greys", 
    edgecolor="black"
)
plt.title("Distribution of Movie Runtime", fontsize=16, weight='bold')
plt.xlabel("Runtime (minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

9. Which date saw the highest movie production?

```python
movies_per_year = df['release_date'].value_counts().sort_index().reset_index()
movies_per_year.columns = ['Date', 'Number of Movies']

# Pick top 10 dates with most movies
top10_years = movies_per_year.sort_values(by='Number of Movies', ascending=False).head(10)

# Display table without gradient
display(top10_years)

# Plot top 10 dates
plt.figure(figsize=(12,6))
sns.barplot(
    data=top10_years.sort_values('Number of Movies'),  # Sort for ascending bars
    x='Date',
    y='Number of Movies',
    color='skyblue'  # Use single color instead of palette to avoid FutureWarning
)
plt.title("Top 10 Dates by Number of Movies Released", fontsize=16, weight='bold')
plt.xlabel("Date")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```

10. Which movies have the longest runtime?

```python
longest_movies = df.sort_values(by="runtime", ascending=False).head(10)

# Display plain table
display(longest_movies[["title", "runtime"]])

# Create horizontal barplot with light red gradient
plt.figure(figsize=(10,6))
colors = sns.color_palette("Reds", n_colors=len(longest_movies))
sns.barplot(
    x="runtime",
    y="title",
    data=longest_movies,
    hue="title",    # Assign y variable to hue for Seaborn 0.14+
    palette=colors,
    dodge=False,
    legend=False
)
plt.title("Top 10 Longest Movies", fontsize=16, weight="bold")
plt.xlabel("Runtime (minutes)", fontsize=12)
plt.ylabel("Movie Title", fontsize=12)
plt.tight_layout()
plt.show()
```

11. Which genres appear most frequently

```python
top_genres = df["genres"].value_counts().head(10).reset_index()

print("Top 10 Genre Movies")

top_genres.columns = ["Genre", "Count"]

# Display table
display(top_genres)

# Plot with Seaborn (fix warning without extra lines)
plt.figure(figsize=(12,6))
sns.barplot(
    x="Genre", 
    y="Count", 
    data=top_genres, 
    hue="Genre",   # Assign hue to fix FutureWarning
    dodge=False,   # Avoid splitting bars
    palette="viridis", 
    legend=False   # Hide legend inline
)

plt.title("Top 10 Most Common Movie Genres", fontsize=16, weight='bold')
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
